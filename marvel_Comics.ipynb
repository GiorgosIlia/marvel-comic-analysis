```python
import os 
import pandas as pd 
import glob
```

```python
df = pd.read_csv("marvel_Comics.csv")
df
```

```python
yolo = df[(df["comic_name"] == "Agents of Atlas (2006 - 2007)")]
yolo
```

```python
df["Year"] = df["publish_date"].str.split(",").str[1]
df
```

```python
unanounced_heroes = df[(df["penciler"] == "None") & (df["writer"] == "None")& (df["cover_artist"] == "None")& (df["Price"] == " Free")]
free = unanounced_heroes.sort_values("publish_date")
hey = free.groupby("Year").count()
hey.drop_duplicates()
hey.drop(hey.iloc[:, 0:11],axis=1,inplace=True,index=None)
hey
```

```python
import matplotlib.pyplot as plt

plt.plot(hey.index,hey["Price"])
plt.ylabel("number of comics")
plt.xlabel("Year")
plt.grid()
plt.title("comics without a writers name or price - unanounced heroes")
ax=plt.gca()
ax.set_xticks(ax.get_xticks()[::5])
plt.xticks(rotation = "vertical",size=8)
plt.show()
```

```python
hey2 = df[df["Imprint"]==" Marvel Universe"]
hey2
```

```python
yol = hey2.groupby("Year").count()
yol
```

```python
plt.bar(yol.index,yol["Price"])
plt.ylabel("number of comics")
plt.xlabel("Year")
plt.grid()
ax=plt.gca()
ax.set_xticks(ax.get_xticks()[::2])
plt.title("comics that Marvel Universe imprinted throughout the years")
plt.xticks(rotation = "vertical",size=8)
plt.show()
```
